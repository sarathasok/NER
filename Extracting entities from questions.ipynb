```python
def extract_named_ents(text):

    return [(ent.text, ent.start_char, ent.end_char, ent.label_) for ent in nlp(text).ents]

def add_named_ents(df):

    df1['named_ents'] = df1['text'].apply(extract_named_ents)
```
